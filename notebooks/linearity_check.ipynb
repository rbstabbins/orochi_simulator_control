# Connect to cameras

```python
import orochi_sim_ctrl as osc

ic = osc.start_ic()
camera_config = osc.load_camera_config()
cameras = osc.connect_cameras(ic, camera_config)
osc.configure_cameras(cameras)
```

```python
for camera in cameras:
    fr = 30.0
    camera.set_frame_rate(fr)
```

# Set ROIs

```python
osc.find_camera_rois(cameras, roi_size=30)
```

# Run script

```python
osc.check_channel_linearity(cameras, n_exp=50)
```
